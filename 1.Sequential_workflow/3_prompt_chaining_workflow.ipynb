```python
from langchain_cohere import ChatCohere
from langgraph.graph import StateGraph,START,END
from dotenv import load_dotenv
from typing import TypedDict
```

```python
class ChainState(TypedDict):
    topic:str
    outline:str
    blog:str
```

```python
load_dotenv()
llm=llm = ChatCohere(model="command-r-plus-08-2024")
```

```python
def write_outline(state:ChainState)->ChainState:
    topic=state["topic"]
    prompt=f"You are a great hardcore writer, so i want you to write a outline for this topic :{topic}"
    state["outline"]=llm.invoke(prompt).content
    return state
```

```python
def write_blog(state:ChainState)->ChainState:
    topic=state["topic"]
    outline=state["outline"]
    prompt=f"you are a very excellent writer so i want you to write a blog on {topic} with the help of this outline:{outline}"
    state["blog"]=llm.invoke(prompt).content
    return state
```

```python
graph=StateGraph(ChainState)

graph.add_node("write_outline",write_outline)
graph.add_node("write_blog",write_blog)

graph.add_edge(START,"write_outline")
graph.add_edge("write_outline","write_blog")
graph.add_edge("write_blog",END)

workflow=graph.compile()
```

```python
output=workflow.invoke({"topic":"Heritages of Nepal"})
```
