```python
from langchain_cohere import ChatCohere
from dotenv import load_dotenv
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
load_dotenv()
```

```python
class LLMState(TypedDict):
    question:str
    answer:str
```

```python
def search_query(state:LLMState)->LLMState:
    query=state["question"]
    llm = ChatCohere(model="command-r-plus-08-2024")
    prompt=f"You are a very helpful assistance so you just answer the question asked by the user in the very simple way,and here is the query:{query}"
    answer=llm.invoke(prompt).content
    state["answer"]=answer
    
    return state
```

```python
graph=StateGraph(LLMState)

graph.add_node("search_query",search_query)

graph.add_edge(START,"search_query")
graph.add_edge("search_query",END)

workflow=graph.compile()
```

```python
workflow.invoke({"question":"What is the capital of Nepal?"})["answer"]
```
